```python
cantidad = 1000000
```

```python
influencer = ["AnaGamer", "LuisGamer", "MarthaGamer", "CarolinaGamer", "SofiaGamer", "PedroGamer"]
likes = [1023,300,1000,400,1977,31]
commentarios = [1,200,300,400,5000,111]
tipo_post = ["Reel", "Publicación", "Reel", "Publicación", "Reel", "Reel"]
```

```python
import random
```

```python
dfEngagement = pd.DataFrame({
    "influencer" : [random.choice(influencer) for _ in range(cantidad)],
    "likes" : [random.choice(likes) for _ in range(cantidad)],
    "commentarios" : [random.choice(commentarios) for _ in range(cantidad)],
    "tipo_post" : [random.choice(tipo_post) for _ in range(cantidad)]
})
```
